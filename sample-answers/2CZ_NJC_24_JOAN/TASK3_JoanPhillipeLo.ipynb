```python
#task 3.1
def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

def task3_1(list_of_values: list):
    if len(list_of_values) <=1:
        return list_of_values
    
    mid = len(list_of_values) //2
    lefthalf = list_of_values[:mid]
    righthalf = list_of_values[mid:]

    sortedleft = task3_1(lefthalf)
    sortedright = task3_1(righthalf)
    
    return merge(sortedleft, sortedright)

task3_1([56,25,4,98,0,18,4,5,7,0])
```

```python
#task 3.2

import random

def task3_2(filename_in:str, filename_out:str, no_of_sample:int):
    with open(f'./Resources/TASK3/{filename_in}') as f:
        data = [line.strip() for line in f]
    numbers =[]
    while len(numbers) < no_of_sample:
        randomindex = random.randint(0, len(data)-1)
        numbers.append(data[randomindex])
    
    with open(f'./Resources/TASK3/{filename_out}', 'w') as f:
        for number in numbers:
            f.write(number + '\n')
    average = sum([float(number) for number in numbers])/len([float(number) for number in numbers])
    filtered = []
    for number in data:
        if float(number) < average:
            filtered.append(float(number))
    return filtered

task3_2('test.txt', 'outputtest.txt', 10)
```
