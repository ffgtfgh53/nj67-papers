```python
#sub-task 1.1

alphabet = [ char for char in 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz']

def task1_1(c: str, n:int):
    if c == ' ':
        return('!')
    else: 
        original_index = alphabet.index(c)
        encrypted_index = original_index + n - 26 if original_index+n>52 else original_index + n
        return alphabet[encrypted_index]

task1_1('T', 13)
```

```python
#task 1.2
def task1_2():
    with open ('./Resources/TASK1/TASK1DATA.txt') as f:
        data = str(f.readline())

    key = ''
    while len(key) < len(data):
        key+='N'
        if len(key) < len(data):
            key+='j'
            if len(key) < len(data):
                key+='c'

    encrypted=[]
    for i,char in enumerate(data):
        try:
            encrypted_char = task1_1(char, alphabet.index(key[i]))
            if char.isupper():
                encrypted_char = encrypted_char.capitalize()
        except:
            encrypted_char = char
        encrypted.append(encrypted_char)

    encrypted = ''.join(encrypted)

    with open('./Resources/TASK1/ENCRYPTEDMESSAGE.txt', 'w') as f:
        f.write(encrypted)
    return encrypted
print(task1_2())
```
