## Task 4.1
The text files `PATIENT.txt`, `STAFF.txt` and `APPOINTMENT.txt` store the comma-separated values for each of the tables in the database.

Write a Python program to read in the data from each file and then store each item of data in the correct place in the database.

Run your program.
<div style="text-align: right">[4]</div>

```python
# Task 4.1
# YOUR CODE HERE
import sqlite3

# Connect to the CLINIC.db database
with sqlite3.connect('Resources/CLINIC.db') as conn:
    cursor = conn.cursor()

    with open('Resources/PATIENT.txt', 'r') as file:
        for line in file:
            data = line.strip().split(',')
            query = f'INSERT INTO PATIENT ("PatientID", "Name", "DOB", "ContactNumber") VALUES (?,?,?,?)'
            cursor.execute(query, data)
    # Close file automatically

    with open('Resources/STAFF.txt', 'r') as file:
        for line in file:
            data = line.strip().split(',')
            query = f"INSERT INTO STAFF ('StaffID', 'Name', 'Role', 'Specialization') VALUES (?,?,?,?)"
            cursor.execute(query, data)
    # Close file automatically

    with open('Resources/APPOINTMENT.txt', 'r') as file:
        for line in file:
            data = line.strip().split(',')
            query = f"INSERT INTO APPOINTMENT ('AppointmentID', 'PatientID', 'StaffID', 'AppointmentDate', 'Diagnosis') VALUES (?,?,?,?,?)"
            cursor.execute(query, data)
    # Close file automatically

    # Commit the transactions and close the connection
    conn.commit()
# Close connection automatically
```

## Task 4.2
Write a Python program that uses an SQL query to retrieve the following data: _Appointment ID_, _Patient ID_, _Patient Name_, _Staff ID_, _Staff Name_, _Appointment Date_, _Diagnosis_ and display them.

Run your program.
<div style="text-align: right;">[5]</div>

```python
# Task 4.2
# YOUR CODE HERE

import sqlite3

with sqlite3.connect('Resources/CLINIC.db') as conn: # Connect to database
    cursor = conn.cursor()              # Create cursor object

    cursor.execute('''
        SELECT Appointment.AppointmentID, Patient.PatientID, Patient.Name, Staff.StaffID, Staff.Name, Appointment.AppointmentDate, Appointment.Diagnosis 
        FROM Appointment
        INNER JOIN Patient ON Appointment.PatientID = Patient.PatientID
        INNER JOIN Staff ON Appointment.StaffID = Staff.StaffID
    ''')
    appointments = cursor.fetchall()  # Fetch all results from SQL query executed

    from pprint import pprint
    pprint(appointments)    # Print appointements in readable format
    print()

# Automatically close database connection
```

## Task 4.3
Write an SQL query to:
* Count the total number of appointments each staff member has handled.
* Group the results by role ( Doctor / Nurse / Technician) and order them by the number of appointments in descending order
* Display the staff members' names, roles and the number of appointments they have handled.

In a new cell, write a Python program that uses this SQL query and display the data under suitable headings

<div style="text-align: right;">[6]</div>

```python
# Task 4.3
# YOUR CODE HERE
import sqlite3

# Connect to the CLINIC.db database
with sqlite3.connect('Resources/CLINIC.db') as conn:
    cursor = conn.cursor()

    # Execute the SQL query to list staff workload
    cursor.execute('''
        SELECT Staff.Name, Staff.Role, COUNT(Appointment.AppointmentID) AS AppointmentCount
        FROM Staff
        LEFT OUTER JOIN Appointment ON Staff.StaffID = Appointment.StaffID
        GROUP BY Staff.StaffID
        ORDER BY Staff.Role, AppointmentCount DESC
    ''')

    # Fetch all the results
    workload = cursor.fetchall()

# Automatically close the database connection

# Print the data under suitable headings
print(f"{'Staff Name':<20} {'Role':<15} {'Number of Appointments':<25}")
print("-" * 60)

for row in workload:
    print(f"{row[0]:<20} {row[1]:<15} {row[2]:<25}")
```
